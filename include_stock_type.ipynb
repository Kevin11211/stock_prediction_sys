```python
import pandas as pd
from talib import abstract

data = pd.read_csv('./test/AAL.csv')
data = data.rename(columns=str.lower)
# 改成 TA-Lib 可以辨識的欄位名稱
selected_data = data[['open','high','low','close','volume']]
```

```python
# 隨意試試看這幾個因子好了
ta_list = ['MACD','RSI','MOM','STOCH']
# 快速計算與整理因子
for x in ta_list:
    output = eval('abstract.'+x+'(selected_data)')
    output.name = x.lower() if type(output) == pd.core.series.Series else None
    selected_data = pd.merge(selected_data, pd.DataFrame(output), left_on = selected_data.index, right_on = output.index)
    selected_data = selected_data.set_index('key_0')
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df = pd.read_csv('./stock_info_all.csv')
```

```python
df.info()
```

```python
df['Type'] = pd.Series()
for i, id in enumerate(df.Symbol):
    ticker = str(id)
    # 获取该代码的Ticker对象
    asset = yf.Ticker(ticker)

    # 获取基本信息
    info = asset.info

    # 判断是否为ETF
    df['Type'][i] = info.get('quoteType')
    print(f"id = {id}, type = {df['Type'][i]}")
```

```python
df.to_csv('stock_all_include_type.csv')
```
