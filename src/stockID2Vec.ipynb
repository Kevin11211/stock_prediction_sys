```python
from gensim.models import Word2Vec
```

```python
import pandas as pd
df = pd.read_csv('./stock_data/stock_all_include_type .csv')
df['Symbol']
```

```python
df.info()
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
maxLength = 0
for e in df['Symbol']:
    maxLength = max(maxLength, len(str(e)))
print(maxLength)
```

```python
# 句子 (也可以是一篇文章)
sentences = [sentence.split() for sentence in df['Symbol'].values]
```

```python
# 設定參數
sg = 0 # sg=1 -> skip-gram, sg=0 -> cbow

# 向前看幾個字或向後看幾個字
window_size = 2 # 越大越好

# 向量維度
vector_size = 5 # 100, 300, 500, ...

# 訓練幾回
epochs = 20

# 最少多少個字才會被使用
min_count = 1

# seed
seed = 42

# 建立 Word2Vec 模型
model = Word2Vec(
    sentences, 
    vector_size=vector_size,  
    window=window_size, 
    sg=sg, 
    min_count=1,
    seed=seed,
    epochs=epochs)
```

```python
# 儲存模型
model.save('stockID2vec.model')

# 讀取模型
model = Word2Vec.load("stockID2vec.model")
```

```python
model.wv['EPR']
```

```python
results = []
for e in df['Symbol']:
    results.append(model.wv[e])

df['Word Embedding'] = results
```
