```python
import pandas as pd

# Manually inputting the data from the image
data = {
    'Category': ['Finance', 'Finance', 'Finance', 'Finance', 'Finance', 
                 'Technology', 'Technology', 'Technology', 'Technology', 'Technology'],
    'Train Accuracy': [0.8213, 0.8261, 0.8067, 0.7831, 0.7562, 0.7488, 0.6638, 0.6526, 0.6411, 0.6631],
    'Train F1 Score': [0.82, 0.83, 0.81, 0.78, 0.76, 0.75, 0.66, 0.65, 0.64, 0.66],
    'Test Accuracy': [0.7727, 0.7899, 0.7726, 0.7517, 0.7027, 0.4902, 0.6344, 0.6433, 0.6261, 0.6081],
    'Test F1 Score': [0.81, 0.82, 0.80, 0.78, 0.75, 0.70, 0.66, 0.65, 0.64, 0.65]
}

df = pd.DataFrame(data)

# Calculating the average values for each category
average_values = df.groupby('Category').mean().reset_index()
```

```python
import matplotlib.pyplot as plt
# Plotting the bar charts for each metric by category
metrics = ['Train Accuracy', 'Train F1 Score', 'Test Accuracy', 'Test F1 Score']
categories = df['Category'].unique()
# Re-plotting the bar charts with single category labels on the x-axis
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

bar_width = 0.35
indices = range(5)

for i, metric in enumerate(metrics):
    finance_values = df[df['Category'] == 'Finance'][metric]
    technology_values = df[df['Category'] == 'Technology'][metric]
    
    axes[i].bar(indices, finance_values, bar_width, label='Finance', alpha=0.7)
    axes[i].bar([index + bar_width for index in indices], technology_values, bar_width, label='Technology', alpha=0.7)
    
    axes[i].set_title(metric)
    axes[i].set_xticks([index + bar_width / 2 for index in indices])
    axes[i].set_xticklabels([f'{index}' for index in indices])
    axes[i].legend()
    axes[i].grid(True)

# Adding single x-axis labels for Finance and Technology
plt.figtext(0.5, -0.01, 'Finance (left) | Technology (right)', ha='center', fontsize=12, bbox={"facecolor":"orange","alpha":0.5,"pad":5})

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
